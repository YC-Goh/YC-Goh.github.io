## Using the Zenodo API

```python
import requests
import pandas as pd
import operator as op
import os
import tempfile
this_nb_name = 'using-zenodo-api'
if not os.path.exists(this_nb_name):
    os.mkdir(this_nb_name)
this_nb_file_path = lambda path: os.path.join(this_nb_name, path)
```

```python
r = requests.get(
    'https://zenodo.org/api/records/5718645'
)
```

```python
if r.status_code == 200:
    file = list(filter(lambda file: file['type'] in ['xls', 'xlsx'], r.json()['files']))[0]
r = requests.get(file['links']['self'], stream=True)
```

```python
this_file_path = this_nb_file_path(file['key'])
if r.status_code == 200:
    if os.path.exists(this_file_path):
        os.remove(this_file_path)
    with open(this_file_path, 'wb') as f:
        for c in r.iter_content(chunk_size=128):
            f.write(c)
```

```python
for n, df in pd.read_excel(this_file_path, sheet_name=None).items():
    print(n, df.shape)
```

```python
fin_df, mkt_df, pnl_df_a, pnl_df_b = op.itemgetter('SreeningFinancial', 'ScreeningMarketInfo', 'Table1-PanelA', 'Table1-PanelB')(pd.read_excel(this_file_path, sheet_name=None))
```
